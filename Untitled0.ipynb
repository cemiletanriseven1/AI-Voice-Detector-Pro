```python
import librosa
import numpy as np

def extract_enhanced_features(file_path):
    try:
        # Sesi yükle ve 5 saniyeye sabitle
        audio, sr = librosa.load(file_path, sr=22050)
        target_len = sr * 5
        if len(audio) > target_len:
            audio = audio[:target_len]
        else:
            audio = np.pad(audio, (0, target_len - len(audio)))

        # MFCC Özellikleri (40 katsayı)
        mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)

        # Özellikleri normalize et
        mfcc = (mfcc - np.mean(mfcc)) / np.std(mfcc)

        return mfcc
    except Exception as e:
        return None
```

```python
import glob

# Tam yollar
KAGGLE_REAL_DIR = os.path.join(KAGGLE_PATH, "deepfake_audio_dataset_jay15k/real")
KAGGLE_FAKE_DIR = os.path.join(KAGGLE_PATH, "deepfake_audio_dataset_jay15k/fake")

def get_files(directory, extensions):
    files = []
    for ext in extensions:
        files.extend(glob.glob(os.path.join(directory, f"*.{ext}")))
    return files

# Dosya listelerini oluşturma
# Kaggle verilerinden ilk 1000'er tanesini alıyoruz (Eğitimi hızlandırmak ve dengelemek için)
kaggle_real_files = get_files(KAGGLE_REAL_DIR, ["wav"])[:1000]
kaggle_fake_files = get_files(KAGGLE_FAKE_DIR, ["wav"])[:1000]

# Senin özel verilerin (Hepsini alıyoruz)
my_real_files = get_files(MY_REAL, ["m4a", "ogg", "wav", "mp3"])
my_fake_files = get_files(MY_FAKE, ["mp3", "wav"])

print(f"Kaggle Real: {len(kaggle_real_files)} | Kaggle Fake: {len(kaggle_fake_files)}")
print(f"Benim Real: {len(my_real_files)} | Benim Fake: {len(my_fake_files)}")
```

```python
from tqdm import tqdm

X = []
y = []

# Verileri işleme fonksiyonu (Daha önce yazdığımız extract_enhanced_features'ı kullanır)
def process_file_list(file_list, label):
    for f in tqdm(file_list):
        feat = extract_enhanced_features(f)
        if feat is not None:
            X.append(feat)
            y.append(label)

print("Gerçek sesler işleniyor...")
process_file_list(kaggle_real_files, 0)
process_file_list(my_real_files, 0)

print("\nSahte sesler işleniyor...")
process_file_list(kaggle_fake_files, 1)
process_file_list(my_fake_files, 1)

# Numpy formatına çevirme
X = np.array(X)
X = X[..., np.newaxis] # CNN için kanal boyutu ekle (40, 200, 1)
y = np.array(y)

print(f"\nFinal Veri Seti Boyutu: {X.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Veriyi %20 test, %80 eğitim olacak şekilde bölüyoruz
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y  # Gerçek/Sahte oranını korur
)

print(f"Eğitim Örneği: {X_train.shape[0]}")
print(f"Test Örneği: {X_test.shape[0]}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization, Dropout, Flatten, Dense, Input

def build_model(input_shape):
    model = Sequential([
        Input(shape=input_shape),

        # 1. Katman: Kenar ve temel frekans takibi
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        # 2. Katman: Daha karmaşık ses dokuları
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        # 3. Katman: Derin özellikler
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4),

        Flatten(),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(1, activation='sigmoid') # 0: Real, 1: Fake
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = build_model(input_shape=X_train.shape[1:])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# En iyi modeli kaydetmek ve aşırı öğrenmeyi (overfitting) engellemek için
callbacks = [
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    ModelCheckpoint('ai_voice_detector_best.h5', save_best_only=True)
]

print("🚀 Model Eğitimi Başlıyor...")
history = model.fit(
    X_train, y_train,
    epochs=50, # Maksimum 50, ama genelde 15-20'de durur
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    shuffle=True
)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# 1. Eğitim Grafiğini Çizdirelim
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Eğitim Başarısı')
plt.plot(history.history['val_accuracy'], label='Test Başarısı')
plt.title('Model Başarımı')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Test Kaybı')
plt.title('Model Kaybı')
plt.legend()
plt.show()

# 2. Karmaşıklık Matrisi (Hangi sınıfta hata yapıyor?)
y_pred = (model.predict(X_test) > 0.5).astype("int32")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek Değer')
plt.title('Karmaşıklık Matrisi')
plt.show()

print(classification_report(y_test, y_pred))
```

```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import librosa
import numpy as np
import os
from audio_recorder_streamlit import audio_recorder

st.set_page_config(page_title="AI Voice Detector Pro", page_icon="🎤")

@st.cache_resource
def load_my_model():
    return tf.keras.models.load_model('/content/drive/MyDrive/audio_ai_project/final_model_v2.h5')

model = load_my_model()

def extract_features(audio_path):
    audio, sr = librosa.load(audio_path, sr=22050)
    target_len = sr * 5
    # Sessizlik temizleme ve padding
    audio, _ = librosa.effects.trim(audio)
    if len(audio) > target_len:
        audio = audio[:target_len]
    else:
        audio = np.pad(audio, (0, target_len - len(audio)))

    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)
    mfcc = (mfcc - np.mean(mfcc)) / (np.std(mfcc) + 1e-9)
    mfcc = mfcc[np.newaxis, ..., np.newaxis]
    return mfcc

st.title("🎤 AI Voice Detector Pro")

if 'audio_path' not in st.session_state:
    st.session_state.audio_path = None

# Arayüz
col1, col2 = st.columns(2)
with col1:
    uploaded_file = st.file_uploader("Ses Yükle", type=['wav', 'mp3', 'm4a'])
    if uploaded_file:
        with open("temp.wav", "wb") as f: f.write(uploaded_file.getbuffer())
        st.session_state.audio_path = "temp.wav"
with col2:
    audio_bytes = audio_recorder(text="Kaydet", pause_threshold=2.0)
    if audio_bytes:
        with open("temp.wav", "wb") as f: f.write(audio_bytes)
        st.session_state.audio_path = "temp.wav"

if st.session_state.audio_path:
    st.audio(st.session_state.audio_path)

    if st.button("🚀 Modeli Test Et"):
        features = extract_features(st.session_state.audio_path)
        prediction = float(model.predict(features)[0][0])
        prediction = max(0.0, min(1.0, prediction))

        st.divider()

        # --- ŞÜPHELİ KATEGORİSİ EKLENMİŞ MANTIK ---
        if prediction > 0.65:
            st.error(f"### 🚨 SONUÇ: YAPAY ZEKA (AI)")
            st.progress(prediction)
            st.write(f"**Yapaylık Analizi:** Model yüksek olasılıkla sentetik tını algıladı (%{prediction*100:.2f})")

        elif 0.40 <= prediction <= 0.65:
            st.warning(f"### ⚠️ ANALİZ EDİLİYOR: İNSAN OLMA İHTİMALİ YÜKSEK")
            st.progress(prediction)
            st.write("**Model Notu:** Ses karakteristiği insan tınısına çok yakın ancak ortam gürültüsü veya kayıt kalitesi nedeniyle düşük seviyeli yapaylık sinyalleri tespit edildi.")
            st.info("Bu durum genellikle mikrofon yankısı veya oda akustiğinden kaynaklanır.")

        else:
            st.success(f"### ✅ SONUÇ: GERÇEK İNSAN")
            st.progress(1.0 - prediction)
            st.write(f"**Doğallık Analizi:** Model insan sesine özgü frekans düzensizliklerini başarıyla doğruladı (%{(1-prediction)*100:.2f})")

        st.divider()

if st.button("🔄 Sıfırla"):
    st.session_state.audio_path = None
    st.rerun()
```
